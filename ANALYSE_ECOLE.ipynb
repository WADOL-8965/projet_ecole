```python
#==========================
#1.IMPORT DES LIBRAIRIES
#==========================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


#=========================
#2.CHARGEMENT DES DONNEES
#=========================
df=pd.read_csv('Ecole.csv')

print('\n===Apercu des donnees===')
print(df.head())


#=========================
#3.EXPLORATION DES DONNEES
#=========================
print('\n===Statistiques descriptives===')
print(df.describe())


#=========================
#4.ANALYSE DES DONNEES
#=========================

#Calcul de la moyenne par eleve
df['moyenne']=df.groupby('Nom_eleve')['Notes'].transform('mean')

#Top 5 des meilleurs eleves
top_eleves=df[['Nom_eleve','moyenne']].drop_duplicates().sort_values(by='moyenne',ascending=False).head(5)
print('\n===Top 5 des meilleurs eleves===')
print(top_eleves.to_string(index=False))

#Meilleur eleve de l'ecole
best=top_eleves.iloc[0]
print('\n===Meilleur eleve===')
print(f"{best['Nom_eleve']} avec {best['moyenne']:.2f}")

#Moyenne par classe
moyenne_par_classe=df.groupby('Classe')['Notes'].mean().sort_values(ascending=False).reset_index()
print('\n===Moyenne par classe===')
print(moyenne_par_classe.to_string(index=False))


#=========================
#5.VISUALISATION DES DONNEES
#=========================

#Graphique Top 5 des meilleurs eleves
sns.set(style='whitegrid',palette='pastel')
plt.figure(figsize=(8,6))
sns.barplot(x='moyenne',y='Nom_eleve',data=top_eleves,palette='viridis')
plt.title('Top 5 des meilleurs eleves')
plt.xlabel('moyenne')
plt.ylabel('eleves')
plt.tight_layout()
plt.savefig('top5_eleve.png')
plt.show()


#Graphique Moyenne par classe
plt.figure(figsize=(6,4))
sns.barplot(x='Classe',y='Notes',data=moyenne_par_classe,palette='coolwarm')
plt.title('moyenne par classe')
plt.xlabel('Classe')
plt.ylabel('Moyenne des notes')
plt.tight_layout()
plt.savefig('best_classe.png')
plt.show()


#Graphique Distribution des notes
plt.figure(figsize=(8,5))
sns.histplot(df['Notes'],bins=10,color='skyblue')
plt.title('Distribution des notes des eleves')
plt.xlabel('Notes')
plt.ylabel("nombre d'eleves")
plt.tight_layout()
plt.savefig('notes.png')
plt.show()


#Graphique Repartition des notes par matieres
plt.figure(figsize=(10,6))
sns.boxplot(x='Matieres',y='Notes',data=df,palette='Set2')
plt.title('Repartition des notes par matieres')
plt.xlabel('Matieres')
plt.ylabel('Notes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('matiere.png')
plt.show()
```
